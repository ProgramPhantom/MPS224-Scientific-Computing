```python
import openpyxl

def create_spreadsheet(filename, strings):
    # Create a new workbook and select the active worksheet
    wb = openpyxl.Workbook()
    ws = wb.active
    
    # Write the headers in the first row
    ws['A1'] = "String"
    ws['B1'] = "Length"
    
    # Write the data starting from row 2
    current_row = 2
    for s in strings:
        ws[f'A{current_row}'] = s
        ws[f'B{current_row}'] = len(s)
        current_row += 1
        
    # Write the footer row with the formula
    ws[f'A{current_row}'] = "Total"
    
    # Calculate the range for the SUM formula (e.g., B2:B4)
    # We subtract 1 from current_row because current_row is now on the "Total" row
    if current_row > 2:
        formula = f"=SUM(B2:B{current_row - 1})"
    else:
        # Fallback just in case the input list was empty
        formula = "=0"
        
    ws[f'B{current_row}'] = formula
    
    # Save the workbook to the specified filename
    wb.save(filename)

# Example usage:
create_spreadsheet('cheese.xlsx', ['Cheddar', 'Wensleydale', 'Stilton', 'Example'])
```
