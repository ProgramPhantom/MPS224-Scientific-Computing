```python
def pisano_sequence(N,n):
    """
    Compute first n Fibonacci numbers mod N 
    """

    if n < 1 or N < 1:
        return False
    if round(n) != n or round(N) != N:
        return False

    # Generate first n fib:
    fib = [0] if n == 1 else [0, 1]

    while len(fib) < n:
        fib.append(fib[-2] + fib[-1])

    pisano = list(map(lambda x: x % N, fib))

    return pisano


def pisano_period(N):
    """
    Find the pisano period modulo N

    We will use the fact that the sequence always starts
    with 0, 1 (0 % n = 0, 1 % n = 1). When we find a new
    0, 1 in the sequence, we have closed the cycle.
    """

    if N < 1:
        return False
    if round(N) != N:
        return False


    fp = 0
    fc = 1

    pp = 0 
    pc = 1

    period = 0
    while True:
        period += 1

        fp, fc = fc, fc+fp

        pp, pc = pc, fc%N

        if pp == 0 and pc == 1:
            return period


N = 6
print(pisano_sequence(N, 120))
print(pisano_period(N))
```
