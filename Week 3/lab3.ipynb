# Task 1

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

xs = np.linspace(0, 1, 1000)  # x coordinates of points on the graph
fs = xs * (1 - xs)                   # y coordinates of points on the graph
gs = xs * (1 - xs) * (1 - 2*xs)

ax.set_title("Two functions")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(xs, fs, label=r"f(x)", linewidth=3)
ax.plot(xs, gs, label=r"g(x)", linewidth=4)

ax.legend()


```

# Task 2

```python
def plot_lissajous(n, m):
    """Plot the lissajous curve"""
    fig, ax = plt.subplots(figsize=(5, 5))

    ax.set_axis_off()

    ts = np.linspace(0, 2*np.pi, 1000)  # x coordinates of points on the graph
    (xs, ys) = (np.sin(n * ts), np.cos(m * ts))

    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    
    ax.plot(xs, ys, color="#aa4a44", linestyle="--")


plot_lissajous(2, 7)
    
```

```python
xs = np.linspace(-11, 11, 1000)
ys = np.linspace(-11, 11, 1000)

x, y = np.meshgrid(xs, ys)


f = x**2 + y**2 + 20*(np.sin(2 * np.pi * x) + np.cos(2 * np.pi * y)) - 100

plt.contour(x, y, f)
```
