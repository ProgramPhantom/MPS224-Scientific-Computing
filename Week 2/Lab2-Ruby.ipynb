```python
from numpy import max, sum, abs, array, zeros
from numpy.linalg import inv
import numpy
```

```python
def hilbert_matrix(n):

	H = zeros((n, n))
	for i in range(n):
		for j in range(n):
			H[i, j] = 1 / (i + j + 1)
	return H

# I imagine Wikipedia states that the entries to the Hilbert matrix are 1/(i+j-1) because it is using 1-indexing while 
# numpy uses 0-indexing (ie, the top left entry of the matrix is (0, 0)). This is standard for computers but not for people.

def row_sum_norm(A):
	"""
	Compute the row sum norm of A
	"""
	return max(sum(abs(A), axis=1))


def condition_number(A):
	"""
	Compute the condition number of A
	"""
	A_inverse = inv(A)

	return row_sum_norm(A) * row_sum_norm(A_inverse)


# Condition number can also be computed using numpy.linalg.cond
```

```python
print(hilbert_matrix(2))
print(row_sum_norm(hilbert_matrix(23)))

print(sum(hilbert_matrix(2), axis=1))

print(condition_number(hilbert_matrix(9)))
print(numpy.linalg.cond(hilbert_matrix(9), p=numpy.inf))
```

```python
def onion(v):
	"""
	Compute the onion matrix using vector v
	"""
	size = len(v) + len(v) - 1

	m = np.empty((size, size), dtype="object")

	layers = range(len(v))
	for l in layers:
		layer_length = size - l*2
		
		for i in range(layer_length):
			m[l, i+l] = v[l]
			m[size-l-1, i+l] = v[l]

	m = np.rot90(m)
	for l in layers:
		layer_length = size - l*2
		
		for i in range(layer_length):
			m[l, i+l] = v[l]
			m[size-l-1, i+l] = v[l]

	return m


print(onion(["a", "b", "c", "d"]))
print(onion(["d", 'e', "d", "h", "s", "b"]))
```
